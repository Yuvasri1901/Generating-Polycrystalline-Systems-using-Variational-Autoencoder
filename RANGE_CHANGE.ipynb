```python
import openpyxl

# Load the Excel workbook
workbook = openpyxl.load_workbook('training.xlsx')

# Select the appropriate worksheets
worksheet = workbook['Sheet1']  # Replace 'Sheet1' with the actual sheet name

# Define the ranges to be swapped
range1 = 'A1:DPE3'
range2 = 'DPF1:IFJ3'
range3 = 'IFK1:MVO3'
range4 = 'MVP1:RLT3'
range5 = 'RLU1:WBY3'

# Read the values from the ranges
values1 = []
values2 = []
values3 = []
values4 = []
values5 = []

for row in worksheet[range1]:
    for cell in row:
        values1.append(cell.value)

for row in worksheet[range2]:
    for cell in row:
        values2.append(cell.value)

for row in worksheet[range3]:
    for cell in row:
        values3.append(cell.value)

for row in worksheet[range4]:
    for cell in row:
        values4.append(cell.value)

for row in worksheet[range5]:
    for cell in row:
        values5.append(cell.value)

# Check the lengths of the ranges
length1 = len(values1)
length2 = len(values2)
length3 = len(values3)
length4 = len(values4)
length5 = len(values5)

# Swap the ranges only if their lengths are the same
if length1 == length2 == length3 == length4 == length5:
    # Write the values to the new ranges
    for index, row in enumerate(worksheet[range2]):
        for cell_index, cell in enumerate(row):
            cell.value = values4[index * len(row) + cell_index]

    for index, row in enumerate(worksheet[range4]):
        for cell_index, cell in enumerate(row):
            cell.value = values2[index * len(row) + cell_index]

    for index, row in enumerate(worksheet[range5]):
        for cell_index, cell in enumerate(row):
            cell.value = values3[index * len(row) + cell_index]

    for index, row in enumerate(worksheet[range3]):
        for cell_index, cell in enumerate(row):
            cell.value = values1[index * len(row) + cell_index]

    for index, row in enumerate(worksheet[range1]):
        for cell_index, cell in enumerate(row):
            cell.value = values5[index * len(row) + cell_index]
else:
    print("Ranges have different lengths. Unable to swap.")

# Save the modified workbook
workbook.save('modified_excel_file.xlsx')
```

```python
import os

def change_image_names(image_files):
    # Get a list of all files in the folder
    file_list = os.listdir(image_files)

    # Filter and sort only the image files
    image_files = sorted([
        filename for filename in file_list
        if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))
    ])

    # Iterate over the image files and rename them sequentially
    for index, old_name in enumerate(image_files):
        # Construct the new name for the file with a sequential number
        new_name = f"cut_imagef58_{index+1}.jpg"

        # Get the full paths of the old and new files
        old_path = os.path.join(folder_path, old_name)
        new_path = os.path.join(folder_path, new_name)

        # Rename the file
        os.rename(old_path, new_path)
        print(f"{old_name} to {new_name}")

# Provide the path to the folder containing the images
folder_path = 'image_files'
change_image_names(folder_path)
```

```python
from PIL import Image

def divide_image(image_files):
    # Open the image
    image = Image.open(image_files)
    
    # Get the width and height of the original image
    width, height = image.size
    
    # Calculate the width and height of each smaller image
    small_width = width // 4
    small_height = height // 2
    
    # Create a list to store the divided images
    divided_images = []
    
    # Iterate through the image and divide it into 8 smaller images
    for i in range(2):
        for j in range(4):
            # Calculate the coordinates of the smaller image
            left = j * small_width
            upper = i * small_height
            right = (j + 1) * small_width
            lower = (i + 1) * small_height
            
            # below code will crop the image from original image
            divided_image = image.crop((left, upper, right, lower))
            
            # Append the divided image to the list
            divided_images.append(divided_image)
    
    return divided_images
#Generating_12_images_size_will_be_correctly_matching
#perfect
# Example usage
image_files = 'image_files/imagef58_22.jpg'
divided_images = divide_image(image_files)

# Save the divided images
for i, divided_image in enumerate(divided_images):
    divided_image.save(f'image_files/cut_imagef68_{i+169}.jpg')
```

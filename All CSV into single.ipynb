```python
import os
import pandas as pd

# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Iterate over the CSV files
for filename in os.listdir('csv_filepath'):  # Replace 'csv_files' with the directory where your CSV files are located
    if filename.endswith('.csv'):
        csv_path = os.path.join('csv_filepath', filename)
        
        # Read each CSV file into a DataFrame
        df = pd.read_csv(csv_path)
        df=df.T
        
        # Append the DataFrame to the combined DataFrame
        combined_df = combined_df.append(df)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combine_imagef_78_8p.csv', index=False)
```
